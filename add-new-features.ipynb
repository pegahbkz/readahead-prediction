```python
import pandas as pd
```

```python
data1 = pd.read_csv("rseq+workload.csv", low_memory=False)
data2 = pd.read_csv("rreverse+workload.csv", low_memory=False)
data3 = pd.read_csv("rrandom+workload.csv", low_memory=False)
data4 = pd.read_csv("rrwr+workload.csv", low_memory=False)
```

```python
data = pd.concat([data1, data2, data3, data4], ignore_index=True)
```

```python
data
```

```python
# Calculate operations per second (ops/sec) for each row
# Assuming 'time_elapsed' is in milliseconds, converting it to seconds for ops calculation
data['ops_per_sec'] = 1 / (data['time_elapsed'] / 1000)

# Calculate mean of 'time_elapsed'
mean_time_elapsed = data['time_elapsed'].mean()

# Calculate distance from mean for each row
data['distance_from_mean'] = data['time_elapsed'] - mean_time_elapsed

# Display the first few rows to verify the new columns
data.head()
```

```python
# Calculate Cumulative Time Elapsed
data['cumulative_time_elapsed'] = data['time_elapsed'].cumsum()

# Calculate Time Difference (using .diff() method, filling NA values with 0 for the first row)
data['time_difference'] = data['time_elapsed'].diff().fillna(0)


# Drop the temporary 'timestamp' column
data.drop(columns=['timestamp'], inplace=True)

# Preview the updated dataframe
data.head()
data = data.drop(['hour', 'minute', 'second', 'millisecond'], axis=1)
# Save the updated dataframe to a new CSV file
data.to_csv("FINAL.csv", index=False)
```
