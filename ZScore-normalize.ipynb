```python
import pandas as pd
```

```python
data = pd.read_csv("FINAL.csv", low_memory=False)
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from scipy.stats import zscore
# Apply Z-score normalization to all numeric columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data_normalized = data.copy()
workload_col = data['workload_type']
data_normalized = data_normalized.drop(['workload_type'], axis=1)
data_normalized[numeric_cols] = data[numeric_cols].apply(zscore)

# Display the first few rows of the normalized dataframe to verify
data_normalized.head()
```

```python
data_normalized['workload_type'] = workload_col
```

```python
data_normalized.info()
```

```python
data_normalized
```

```python
data_normalized = data_normalized.sample(frac = 1)
```

```python
data_normalized
```

```python
data_normalized.to_csv("FINALFINAL.csv", index=False)
```

```python
data_sampled1 = data_normalized.sample(frac=0.01, random_state=42)
```

```python
data_sampled1.to_csv("sampled-data1.csv", index=False)
```
