```python
import pandas as pd
```

```python
data = pd.read_csv("sampled-data1.csv", low_memory=False)
```

```python
data.shape
```

```python
from sklearn.manifold import TSNE
```

```python
modell = TSNE(learning_rate=50)
```

```python
tsne_features = modell.fit_transform(data)
```

```python
tsne_features[1:10,:]
```

```python
data['x'] = tsne_features[:,0]
data['y'] = tsne_features[:,1]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(x="x", y="y", data=data)
plt.show()
```
